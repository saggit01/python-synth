```python
#type nums: List[int]

nums = [0, 1, 2, 3, 4, 5]
r = len(nums) - 1
l = 0
target = 5

class Solution:
    def __init__(self, nums, l, r, target):
        self.nums = nums
        self.l = l
        self.r = r
        self.target = target
        
    def binarySearch(self, nums, l, r, target):
      #r = len(nums) - 1
      #l = 0
      while (l <= r):
        m = int(l + (r - l)/2)
        
        if nums[m] == target:
          return m
        if nums[m] < target:
          l = m + 1
        elif nums[m] > target:
          r = m - 1
      return -1

sol = Solution(nums, l, r, target)
result = sol.binarySearch(nums, l, r, target)
print("Result:", result)
```
